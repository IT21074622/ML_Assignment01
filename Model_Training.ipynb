```python
#Improting Libraries
import pandas as pd
import numpy as np
```

```python
#Importing the files and assign to dataframe
dftrain = pd.read_csv('/content/drive/MyDrive/ML project/preprocessed_train_new.csv')
dftest = pd.read_csv('/content/drive/MyDrive/ML project/preprocessed_test_new.csv')
```

```python
# Reorder dftest columns to match dftrain
dftest = dftest[dftrain.columns]
```

```python
print(dftrain.columns.equals(dftest.columns))  # Should print True
```

```python
dftrain.head()
```

```python
dftest.head()
```

```python
x_train=dftrain.drop('loan_status',axis=1)
y_train=dftrain['loan_status']
x_test=dftest.drop('loan_status',axis=1)
y_test=dftest['loan_status']
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
!pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE

# Create SMOTE object
smote = SMOTE(sampling_strategy=0.67, random_state=42)

# Apply only on training set
x_train_bal, y_train_bal = smote.fit_resample(x_train, y_train)
```

```python
from collections import Counter

print("Before SMOTE:", Counter(y_train))
print("After SMOTE:", Counter(y_train_bal))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
model = LogisticRegression(penalty='l2', C=0.6, solver='lbfgs', max_iter=1000, random_state=42)
model.fit(x_train_bal, y_train_bal)
```

```python
y_pred = model.predict(x_test)
```

```python
# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Classification report
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, precision_recall_curve, average_precision_score
import numpy as np

# Compute normalized confusion matrix
cm = confusion_matrix(y_test, y_pred, normalize='true')  # row-wise normalization

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt=".2f", cmap="Blues", xticklabels=[0, 1], yticklabels=[0, 1])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Normalized Confusion Matrix")
plt.show()

# Precision and Recal
y_scores = model.predict_proba(x_test)[:, 1]
precision, recall, thresholds = precision_recall_curve(y_test, y_scores)
avg_precision = average_precision_score(y_test, y_scores)
plt.figure(figsize=(8, 6))
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label=f'Avg Precision = {avg_precision:.2f}', linewidth=2)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.grid(True)
plt.show()

```
